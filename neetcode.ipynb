```python
# kth largest integer
from heapq import nlargest, heapify, heappop, heappush
from typing import List

# core idea: if we keep a min heap of size k
# then the kth largest is always the root
class KthLargest:
    def __init__(self, k: int, nums: List[int]):
        self.k = k
        self.h = nlargest(self.k, nums)
        heapify(self.h)

    def add(self, val: int) -> int:
        if len(self.h) == 0:
            heappush(self.h, val)
        elif val > self.h[0]: # case: need to add to heap
            heappop(self.h) # discard previous kth largest
            heappush(self.h, val)
        return self.h[0] # always return kth_largest

["KthLargest", [1, []], "add", [3], "add", [-2], "add", [5], "add", [10], "add", [9]]
kth_largest = KthLargest(1, [])
print ( "expect 3")
print (kth_largest.add(3))
```

```python
# Binary search
class BinarySearch:
    def search(self, nums: List[int], target: int) -> int:
        # [5]
        def _search(left, right):
            if left > right:
                return -1
            guess_index = (left + right) // 2
            guess_val = nums[guess_index]
            if guess_val == target:
                return guess_index
            if guess_val < target:
                return _search(guess_index + 1, right)
            return _search(left, guess_index - 1)
        return _search(0,len(nums)-1)

    def iterative_search(self, nums, target):
        left, right = 0, len(nums) - 1
        while left <= right:
            mid = (left + right) // 2
            if nums[mid] == target:
                return mid
            if nums[mid] < target:
                left = mid + 1
            else:
                right = mid - 1
        return -1

b = BinarySearch()
print (b.search([-1,0,2,4,6,8],4))
print ("expect: 3")
```

```python
# Validate Parentheses
from collections import deque

class Solution:
    def isValid(self, s: str) -> bool:
        stack = deque()
        matches = {"(": ")", "[": "]", "{": "}"}
        for c in s:
            if c in "([{":
                stack.append(c)
            else:
                if len(stack) == 0:
                    return False

                popped = stack.pop()
                if c == matches[popped]:
                    continue
                else:
                    return False
        return (len(stack) == 0)

class _Solution:
    def isValid(self, s: str) -> bool:
        stack = deque()
        matches = {"(": ")", "[": "]", "{": "}"}
        for c in s:
            if c in matches:
                stack.append(c)
            elif len(stack) == 0 or c != matches[stack.pop()]:
                return False

        return len(stack) == 0

# learned empty data structures are "falsy"
class GrokSolution:
    def isValid(self, s: str) -> bool:
        stack = []
        matches = {"(": ")", "[": "]", "{": "}"}
        for c in s:
            if c in matches:
                stack.append(c)
            elif not stack or matches.get(stack.pop()) != c:
                return False
        return not stack

```

```python
# best time to buy and sell stock
class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        max_profit = 0
        lowest_ask = 0
        for idx, price in enumerate(prices):
            if idx == 0:
                lowest_ask = price
                continue
            
            if price < lowest_ask:
                lowest_ask = price
            this_profit = price - lowest_ask
            if this_profit > max_profit:
                max_profit = this_profit
        return max_profit

#lmao, literally forgot about min and max functions, this is better
class _Solution:
    def maxProfit(self, prices: List[int]) -> int:
        max_profit = 0
        min_price = float('inf')
        for price in prices:
            min_price = min(min_price, price)
            max_profit = max(max_profit, price - min_price)
        return max_profit

# learning functools. I think I need to implement that library to remember what they all do
class FunctionalSolution:
    def maxProfit(self, prices: List[int]) -> int:
    return reduce(
        lambda acc, current: (min(acc[0], current), max(acc[1], current - acc[0])),
        prices,
        (float('inf'), 0)
    )[1]
```

```python
class Solution:
    def isPalindrome(self, s: str) -> bool:
        def reverse(_s):
            return _s[::-1]
        stripped = ''.join([c for c in s.lower() if c.isalnum()])
        return stripped == reverse(stripped)
```

```python
# valid anagram
# first attempt:
class First:
    def isAnagram(self, s: str, t: str) -> bool:
        if len(s) != len(t):
            return False
        d = {}
        for idx, char in enumerate(s):
            # add for s
            if char in d:
                d[char] = d[char] + 1
            else:
                d[char] = 1
            # remove for t
            subchar = t[idx]
            if subchar in d:
                d[subchar] = d[subchar] - 1
            else:
                d[subchar] = -1
        
        for value in d.values():
            if value == 0:
                continue
            else:
                return False
        return True

# second attempt. realized you could compare maps with ==
class Second:
    def isAnagram(self, s: str, t: str) -> bool:
        if (len(s) != len(t)):
            return False
        _s = {}
        _t = {}
        for i in range(len(s)):
            if s[i] in _s:
                _s[s[i]] = _s[s[i]] + 1
            else:
                _s[s[i]] = 1
            if t[i] in _t:
                _t[t[i]] = _t[t[i]] + 1
            else:
                _t[t[i]] = 1
        
        return _s == _t

# third attempt, learned the get method accepts a default value to use instead of None
class Solution:
    def isAnagram(self, s: str, t: str) -> bool:
        if (len(s) != len(t)):
            return False

        def count(s):
            counts = {}
            for c in s:
                counts[c] = counts.get(c, 0) + 1
            return counts

        _s = count(s)
        _t = count(t)
        
        return _s == _t
```
